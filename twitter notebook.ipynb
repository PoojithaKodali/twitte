```python
for i in range(0, len(data)):
     # 3a. Remove the numbers and text
    data['Tweet'][i] = data['Tweet'][i].split('http')[0] 
    
    
    #  3c. Remove text before ":" (colon)
    if data['Tweet'].loc[i].find(':') != -1:
        data['Tweet'].loc[i] = data['Tweet'].loc[i].split(':')[1]
        
    #  3d. Remove all hastags (start with #)
    data['Tweet'].loc[i] = ' '.join(x for x in data['Tweet'][i].split() if not x.startswith('#'))

    #  3e. Remove all usernames (start with @)
    data['Tweet'].loc[i] = ' '.join(x for x in data['Tweet'][i].split() if not x.startswith('@'))
```

```python
for i in range(0,len(data)):
    data['Tweet'][i]=re.sub(r'[a-zA-Z]', '',data['Tweet'][i])
    data['Tweet'][i]=re.sub(r'[0-9]', '',data['Tweet'][i])
```
