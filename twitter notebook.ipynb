```python
import pandas as pd
import re
import nltk
```

```python
data=pd.read_csv(r'C:\Users\Poojitha\Documents\twitter\tweets.csv')
```

```python
data.head
```

```python
for i in range(0, len(data)):
     # 3a. Remove the numbers and text
    data['Tweet'][i] = data['Tweet'][i].split('http')[0] 
    
    
    #  3c. Remove text before ":" (colon)
    if data['Tweet'].loc[i].find(':') != -1:
        data['Tweet'].loc[i] = data['Tweet'].loc[i].split(':')[1]
        
    #  3d. Remove all hastags (start with #)
    data['Tweet'].loc[i] = ' '.join(x for x in data['Tweet'][i].split() if not x.startswith('#'))

    #  3e. Remove all usernames (start with @)
    data['Tweet'].loc[i] = ' '.join(x for x in data['Tweet'][i].split() if not x.startswith('@'))
```

```python
for i in range(0,len(data)):
    data['Tweet'][i]=re.sub(r'[a-zA-Z]', '',data['Tweet'][i])
    data['Tweet'][i]=re.sub(r'[0-9]', '',data['Tweet'][i])
```

```python
  special_symbols = "\"|\;|\-|\''|\\|\/|\/|\"|\'|\?|\<|\>|\?|\:|\||\!|\~|\@|\#|\$|\%|\^|\`|\&|\*|\(|\)|\_|\+|\‚Äì|\=|\{|\}|\[|\]|\.|\,"


data['Tweet'].replace(to_replace = special_symbols, value = '', regex = True, inplace = True)
```

```python

for i in range(len(data)):
    data['Tweet'][i] = data['Tweet'][i].replace("‚Äú", "")
    data['Tweet'][i] = data['Tweet'][i].replace("‚Äù", "")
    data['Tweet'][i] = data['Tweet'][i].replace("‚Äò", "")
    data['Tweet'][i] = data['Tweet'][i].replace("‚Äô", "")
    data['Tweet'][i] = data['Tweet'][i].replace("‚Çπ", "")  
    data['Tweet'][i] = data['Tweet'][i].replace("‚Ä¶‚Ä¶", "")
    data['Tweet'][i] = data['Tweet'][i].replace("‚Ä¶", "")
```

```python
import stanza
```

```python
stanza.download('te')
```

```python
nlp = stanza.Pipeline('te', processors='tokenize,pos')
dic={}
for i in data['Tweet']:
    en_test_= i
    doc = nlp(en_test_)
    sent_list = [sent.text for sent in doc.sentences]
    for sent in doc.sentences:
        for word in sent.words:
            dic[word.text]=word.xpos
print(dic)
```

```python
from nltk.tokenize.casual import TweetTokenizer
t = TweetTokenizer()
output=[]
for i in data['Tweet']:
    words = t.tokenize(i)
    for i in range(len(words)- 3+1):
        trigrams=words[i:i+3]
        dictpos=[]
        for i in trigrams:
            if(i in dic):
                dictpos.append((i,dic[i]))
            else:
                 dictpos.append((i,'PUNCT'))
        output.append(dictpos)
print(output)
        
    

            
```
