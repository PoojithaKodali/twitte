```python
for i in range(0, len(data)):
     # 3a. Remove the numbers and text
    data['Tweet'][i] = data['Tweet'][i].split('http')[0] 
    
    
    #  3c. Remove text before ":" (colon)
    if data['Tweet'].loc[i].find(':') != -1:
        data['Tweet'].loc[i] = data['Tweet'].loc[i].split(':')[1]
        
    #  3d. Remove all hastags (start with #)
    data['Tweet'].loc[i] = ' '.join(x for x in data['Tweet'][i].split() if not x.startswith('#'))

    #  3e. Remove all usernames (start with @)
    data['Tweet'].loc[i] = ' '.join(x for x in data['Tweet'][i].split() if not x.startswith('@'))
```

```python
for i in range(0,len(data)):
    data['Tweet'][i]=re.sub(r'[a-zA-Z]', '',data['Tweet'][i])
    data['Tweet'][i]=re.sub(r'[0-9]', '',data['Tweet'][i])
```

```python
nlp = stanza.Pipeline('te', processors='tokenize,pos')
dic={}
for i in data['Tweet']:
    en_test_= i
    doc = nlp(en_test_)
    sent_list = [sent.text for sent in doc.sentences]
    for sent in doc.sentences:
        for word in sent.words:
            if word.pos=='NOUN' or word.pos=='ADV' or word.pos=='ADJ' or word.pos=='VERB':
                dic[word.text]=word.xpos
            else:
                pass
print(dic)
```

```python
from nltk.tokenize.casual import TweetTokenizer
t = TweetTokenizer()
trigrams=[]
for i in data['Tweet']:
    words = t.tokenize(i)
    out=[]
    for i in range(len(words)- 3+1):
        trigram=words[i:i+3]
        tri=[]
        for i in trigram:
            if i in dic:
                tri.append(i)
        out.append(tri)
    trigrams.append(out)
print(trigrams)
```

```python
from googletrans import Translator
from textblob import TextBlob

translator = Translator()
for i in trigrams:
    sentence_nu_c=0
    sentence_p_c=0
    sentence_ne_c=0
    for j in i:
        neutral_count=0
        positive_count=0
        negative_count=0
        for k in j:
            t1_text=translator.translate(k)
            english_word1=t1_text.text
            eng_text=TextBlob(english_word1)
            res=eng_text.sentiment.polarity
            if(res<0):
                negative_count+=1
            elif(res>0):
                positive_count+=1
            else:
                neutral_count+=1
        if(positive_count>negative_count):
            sentence_p_c+=1
        elif(negative_count>positive_count):
            sentence_ne_c+=1
        else:
            sentence_nu_c+=1
            
    
    if(sentence_ne_c<sentence_p_c):
        print("POSITIVE")
    elif(sentence_p_c<sentence_ne_c):
        print("NEGATIVE")
    else:
        print("NEUTRAl")
            
   

    
    
```
